9) Check number of unique values in each column, can they be combined in any way to ease encoding? Comment your thoughts and make those changes.

```python
categorical_df = customer_df.select_dtypes(object)

for col in categorical_df:
    display(categorical_df[col].value_counts())
```

```python
categorical_df['State'] = categorical_df['State'].replace({'Arizona': 'Other State',
                                                           'Nevada':'Other State', 'Washington':'Other State'})
```

Since Arizona, Nevada and Washington are on the low side of the spectrum I decided to merge them all as 'Other State'

```python
categorical_df['Education'] = categorical_df['Education'].replace({'Master':'Postgrad','Doctor':'Postgrad'})
```

Same situation as before with 'Master' and 'Doctor', since they are both relatively low I decided to combine them both as 'Postgrad'

```python
categorical_df['EmploymentStatus'] = categorical_df['EmploymentStatus'].replace({'Medical Leave':'Not looking for work',
                                                                                 'Disabled':'Not looking for work',
                                                                                 'Retired':'Not looking for work'})
```

I turned this one in a 3 way category, differentiating which customers are working, unemployeed and not looking for work.

```python
categorical_df['Marital Status'] = categorical_df['Marital Status'].replace({'Divorced':'Single'})
```

For the marital status I didn't think it was relevant to know which customers are divorced so I joined them with the singles.

```python
categorical_df['Vehicle Class'] = categorical_df['Vehicle Class'].rename({'Sports Car':'High end car','Luxury SUV':'High end car', 
                                                                          'Luxury Car': 'High end car'})
```

And finally for Vehicle Class I combined the luxury and sports car categories into one 'high end car'.
